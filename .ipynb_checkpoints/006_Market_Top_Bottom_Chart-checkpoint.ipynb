- 1970 년대 이후의 다우 존수 종합 지수 변동 시각화

```python
dji = pd.read_csv('data/DJI_1970.csv', index_col=0, parse_dates=True)
dji.head(2)
```

```python
import re
usbond = pd.read_csv("USA10yearBond.csv")
usbond['날짜'] = usbond['날짜'].apply(lambda x: re.sub('년 ', '/', x))
usbond['날짜'] = usbond['날짜'].apply(lambda x: re.sub('월', '/1', x))
usbond['날짜'] = pd.to_datetime(usbond['날짜'])
usbond.set_index(usbond['날짜'], inplace=True)
usbond.head(2)
```

```python
dji['Close'].plot()
plt.title("다우존스 산업 지수")
```

```python
usbond['종가'].plot()
plt.title("미국 10년물 국채 금리")
```

### 1973~74년 1차 오일 쇼크
- 1979년 이란 혁명. 전세계적으로 물가는 상승하면서 실업 등의 문제는 오히려 심각해지는 스태그플레이션이 나타났으며, 미국의 FRB는 급작스럽게 불어난 달러를 미국 은행으로 회수하기 위해 금리를 21%까지 인상

```python
series = dji.loc['1970-01-02':'1975-12-31']['Close']
series.plot()
```

```python
print("최고점", series.idxmax(), series.max())
print("최저점", series.idxmin(), series.min())
print("Drawdown {:.1f}%".format((series.max() - series.min())/series.max()*100))
```

### 2000 년 3월 dot-com bubble의 Nasdaq 붕괴
1) 3/7 거래량이 크게 증가하며 하락  
2) 3/10 오전장에서만 85 포인트 뛰어 오르며 사상 최고치 경신. 오후에 급락하여 불과 2 포인트 상승으로 마감.  거래량은 평균치 보다 13%나 많음. stall day (기관이 팔고 있다는 표시)  
3) 3/14 대량 거래와 함께 4%나 하락  
4) 3/16~24 까지 상승세. 랠리가 다시 시작되는 것 아닌가 하는 착각을 불러 일으킴.  
5) 4번째 매물 출회되며 상승세에 찬물을 끼얹음.  stall day.  
6) 5번째 매물 출회

```python
df = yf.download('^IXIC', start='1999-01-01', end='2002-12-31', progress=False)
df.head()
```

```python
series = df['Close']
series.plot()
```

```python
print("최고점", series.idxmax(), series.max())
print("최저점", series.idxmin(), series.min())
print("Drawdown {:.1f}%".format((series.max() - series.min())/series.max()*100))
```
