# Portfolio Values and Sharp Ratio

- Apple, Cisco, IBM, Amazon 으로 포트폴리오 구성  
- 최초 구성한 비율을 끝까지 유지한다고 가정하여 portpolio value 분석

```python
import yfinance as yf
import datetime
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2017, 1, 1)
```

```python
aapl_t = yf.Ticker('AAPL')
aapl = aapl_t.history(start=start, end=end)
aapl.head()
```

```python
cisco_t = yf.Ticker('CSCO')
ibm_t = yf.Ticker('IBM')
amzn_t = yf.Ticker('AMZN')
```

```python
cisco = cisco_t.history(start=start, end=end)
cisco.head()
```

```python
ibm = ibm_t.history(start=start, end=end)
ibm.head()
```

```python
amzn = amzn_t.history(start=start, end=end)
amzn.head()
```

## Normalized daily Close 계산

```python
for df in [aapl, cisco, ibm, amzn]:
    df['normed Close'] = df['Close'] / df['Close'][0]
```

```python
aapl.head()
```

## Portfolio 구성

- 다음과 같이 portfolio 를 구성했다고 가정
    - Apple : 30%
    - Cisco : 20%
    - Amazon : 40%
    - IBM : 10%
    
    
- 구성비율을 Normed Return 에 곱하여 각 종목의 weighted normed return 계산

```python
for df, alloc in zip([aapl, cisco, amzn, ibm], [.3, .2, .4, .1]):
    df['allocation'] = df['normed Close'] * alloc
```

```python
aapl.head()
```

## fund 의 초기 금액이 1000000 이라 가정하고 portfolio 의 종목별 position 계산

```python
for df in [aapl, cisco, amzn, ibm]:
    df['position'] = df['allocation'] * 1000000
```

```python
aapl.head()
```

## 각 종목의 position value 로 구성된 portfolio value  DataFrame 을 생성

```python
port_vals = pd.concat([aapl['position'], cisco['position'], amzn['position'], ibm['position']], axis=1)
port_vals.columns = ['Apple', 'Cisco', 'Amazon', 'IBM']
port_vals['Total Position'] = port_vals.sum(axis=1)
```

```python
port_vals.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
port_vals['Total Position'].plot(ax=ax, label="Total Portfolio Value")
port_vals.drop('Total Position', axis=1).plot(ax=ax)
plt.legend()
```

## Portfolio Statistics

- daily return  
- cumulative return  
- average daily return  
- std of daily return  
- Sharpe Ratio

### Porfolio 의 일일수익률

```python
port_vals['Daily Return'] = port_vals['Total Position'].pct_change(1)
port_vals.head()
```

### 누적수익률

- 누적수익률=(1+당일수익률)∗전일자누적수익률


- 누적수익률 = (Last Total Position) / (First Total Position)

```python
port_vals['CumReturn'] =  (1 + port_vals['Daily Return']).cumprod() - 1
port_vals.tail()
```

```python
cum_ret = port_vals['Total Position'][-1] / port_vals['Total Position'][0] - 1 
print("Cumulative return : {:.2f}%".format(cum_ret * 100))
```

## Sharpe Ratio

- Sharpe ratio = (Mean portfolio return - Risk free rate) / Standard deviaion of portfolio return


- (기대수익률 - 무위험수익률) / (기대수익률 - 무위험수익률의 표준편차)  

- Annualized Sharpe Ratio = K-value * SR  


- K-values
    - Daity - sqrt(252)
    - Weekly - sqrt(52)
    - Monthly - sqrt(12)  
    
    
- thumb of rules
    - ASR > 1 : good
    - ASR > 2 : very good
    - ARR > 3 : excellent

```python
sr = port_vals['Daily Return'].mean() / port_vals['Daily Return'].std()   # risk free rate = 0 로 가정
sr
```

annualized SR

```python
asr = (255 ** 0.5) * sr; asr
```
