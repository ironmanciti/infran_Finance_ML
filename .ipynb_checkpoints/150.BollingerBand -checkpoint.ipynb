# Bollinger Band 
- 1980년대 존 볼린저가 개발한 변동성 지표
- SMA (Simple Moving Average)를 중심으로 상하 m sigma line (수정 가능) 사이의 구간  

- Standard deviation 은 변동성(volatility)의 척도이므로, 시장의 변동성이 커지면 band가 넓어지고, 변동성이 작아지면 band 가 축소된다.  

    - 상한선=MA(TP, n) + m ∗ sigma(TP, n)
    - 하한선=MA(TP, n) − m ∗ sigma(TP, n)
    - n: n일 이동평균, TP(Typical Price): (고가+저가+종가)/3,  m: 보통 2
    
    
- 밴드폭의 변화와 밴드 이탈 여부 등 주가의 변동성으로 미래 주가를 예측  

   - 전일 주가가 전일 하단 밴드보다 높고 현재 주가가 당일 하단 밴드보다 낮으면(price가 lower Bollinger Band를 하향 돌파) 매수 signal 

   - 전일 주가가 전일 상단 밴드보다 낮고 현재 주가가 당일 상단 밴드보다 높으면(price가 upper Bollinger Band를 상향 돌파) 매도 signal

```python
import yfinance as yf
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2022, 6, 1)
```

```python
stock = yf.Ticker('323410.KS')  #카카오뱅크
stock = stock.history(start=start, end=end)
stock.tail()
```

```python
window = 20
m = 2

# 이동평균
mu = stock['Close'].rolling(window).mean()
# 변동성
std = stock['Close'].rolling(window).std()

# Upper / Lower Band
upper_band = mu + m * std
lower_band = mu - m * std
```

- 볼린저밴드 시각화

```python
ax = stock['Close'].plot(title='Bollinger Band', label='Stock Price', figsize=(12, 8))

mu.plot(label=f'{window} day Rolling Mean', ax=ax)
upper_band.plot(label='Upper Band', ax=ax)
lower_band.plot(label='Lower Band', ax=ax)

ax.set_ylabel("Price")
ax.legend()
```
