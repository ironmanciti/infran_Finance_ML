```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Augmented Dickey Fuller test

- 육안으로 time series 의 정상성 (stationary) 여부를 판단하는 것은 어려움  


- statsmodel 의 Time Series Analysis (tsa) library 사용  


- statsmodels.tsa 는 time series analysis 에 유용한 함수들 제공


- ADF Test
    - https://en.wikipedia.org/wiki/Augmented_Dickey%E2%80%93Fuller_test
    
Augmented Dickey-Fuller의 귀무가설은 단위근이 있다는 것이고 대안은 단위근이 없다는 것입니다. p값이 임계 크기보다 크면 단위 루트가 있다는 것을 거부할 수 없으므로 time series는 non-stationary 합니다.

```python
from statsmodels.tsa.stattools import adfuller

def adf_check(ts):
    result = adfuller(ts)
    print('p-value: {:f} - {}'.format(result[1], 
                        'Non-stationary' if result[1] > 0.05 else 'Stationary'))
```

```python
df = pd.read_csv('data/international-airline-passengers.csv', header=0, index_col=0)
df.head()
```

```python
X = df.values
X.shape
```

```python
plt.plot(X)
plt.xticks(rotation=45);
```

```python
adf_check(X)
```

- p-value > 0.05 이므로 H0 (non-stationary)를 기각할 수 없음 -> H0(non-stationary) 채택

# Trend 와 Seasonality

## 단순 trends 만 있는 Simple Time Series

```python
time = np.arange(4 * 365 + 1)
time
```

```python
series = 0.1 * time

plt.figure(figsize=(8, 4))
plt.plot(time, series)
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

```python
adf_check(series)
```

## Seasonal pattern 을 가진 Time Series

- 일정 기간별로 동일한 pattern 반복

```python
def seasonal_pattern(season_time):
    return np.where(season_time < 0.4,
                    np.cos(season_time * 2 * np.pi),  # True
                    1 / np.exp(3 * season_time))      # False

def seasonality(time, period, amplitude=1, phase=0):
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

**예를 들어, 마이너스 매출로 마감된 다음날 약간의 플러스로 회복하고 주간 동안 줄어들다가 주말에 다시 매출이 급격히 늘어나는 패턴**  

```python
baseline = 10
amplitude = 40
series = seasonality(time, period=365, amplitude=amplitude)

plt.figure(figsize=(8, 4))
plt.plot(time, series)
plt.show()
```

```python
adf_check(series)
```

- p-value < 0.05 이므로 H0 (non-stationary) 기각 -> H1(stationary) 채택

## trend 와 seasonality 모두 가진 Time Series 

```python
slope = 0.05
series = baseline + slope * time + \
            seasonality(time, period=365, amplitude=amplitude)

plt.figure(figsize=(8, 4))
plt.plot(time, series)
plt.show()
```

```python
adf_check(series)
```

# Noise

## white noise 생성

```python
def white_noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
noise_level = 5
noise = white_noise(time, noise_level, seed=42)

plt.figure(figsize=(8, 4))
plt.plot(time, noise)
plt.show()
```

```python
adf_check(noise)
```

## white noise 가 추가된  time series 

```python
series += noise

plt.figure(figsize=(8, 4))
plt.plot(time, series)
plt.show()
```

```python
adf_check(series)
```

## 자기상관성 부여

- 특정 시점 (step) 이 다른 시점 (step) 과 수치적 연관성 있음

```python
def autocorrelation(time, amplitude, seed=None):
    rnd = np.random.RandomState(seed)
    lo = 0.8
    ar = rnd.randn(len(time) + 1)
    for step in range(1, len(time) + 1):
        ar[step] += lo * ar[step - 1]
    return ar[1:] * amplitude
```

## autocorrelation

```python
series = autocorrelation(time, 10, seed=42)
plt.plot(time[:200], series[:200])
plt.show()
```

```python
adf_check(series)
```

## autocorrelation + trend

```python
series = autocorrelation(time, 10, seed=42) + 2 * time
plt.plot(time[:200], series[:200])
plt.show()
```

```python
adf_check(series)
```

## autocorrelation + seasonality + trend

```python
series = autocorrelation(time, 10, seed=42) + \
        seasonality(time, period=50, amplitude=150) + 2*time
plt.plot(time[:200], series[:200])
plt.show()
```

```python
adf_check(series)
```

## seasonality 와 trend 의 변화

- time step 200 부근에서 seasonality 와 trend 의 변화가 생긴 경우  

- ex) 금융위기

```python
series = autocorrelation(time, 10, seed=42) + \
        seasonality(time, period=50, amplitude=150) + 2*time
series2 = autocorrelation(time, 5, seed=42) + \
        seasonality(time, period=50, amplitude=2) - time + 550
series[200:] = series2[200:]
plt.plot(time[:300], series[:300])
plt.show()
```

```python
adf_check(series)
```

## 급격한 peak 추가

```python
def impulses(time, num_impulses, amplitude=1, seed=None):
    rnd = np.random.RandomState(seed)
    impulse_indices = rnd.randint(len(time), size=10)
    series = np.zeros(len(time))
    for index in impulse_indices:
        series[index] += rnd.rand() * amplitude
    return series    
```

```python
series = impulses(time, 10, seed=42)
plt.plot(time, series)
plt.show()
```

```python
adf_check(series)
```

## autocorrelation + impulse

- peak 후에 autocorrelate 되어 감소하다 다른 peak 를 만난 경우

```python
def autocorrelation(source, los):
    ar = source.copy()
    max_lag = len(los)
    for step, value in enumerate(source):
        for lag, lo in los.items():
            if step - lag > 0:
                ar[step] += lo * ar[step - lag]
    return ar
```

```python
signal = impulses(time, 10, seed=42)
series = autocorrelation(signal, {1: 0.99})
plt.plot(time, series + signal)
plt.grid()
plt.show()
```

```python
adf_check(series + signal)
```

```python
signal = impulses(time, 10, seed=42)
series = autocorrelation(signal, {1: 0.70, 50: 0.2})
plt.plot(time, series + signal)
plt.grid()
plt.show()
```

```python
adf_check(series + signal)
```

## 태양 흑점 data 의 계절성, 자기 상관성 분석

- 1749 년 부터 매월말 흑점 갯수를 기록한 data  


- pandas asfreq - TimeSeries 를 특정한 frequency 로 표시

    - Week	'W'	one week  
    - WeekOfMonth	'WOM'	  
    - MonthEnd	'M'	calendar month end
    - QuarterEnd	'Q'	calendar quarter end  
    - YearEnd	'A'	calendar year end  
    - Day	'D'	one absolute day
    - Hour	'H'	one hour

```python
df = pd.read_csv("data/sunspots.csv", parse_dates=["Date"], index_col="Date")
series = df["Monthly Mean Total Sunspot Number"].asfreq("1M")
series.head()
```

```python
adf_check(series.values)
```
