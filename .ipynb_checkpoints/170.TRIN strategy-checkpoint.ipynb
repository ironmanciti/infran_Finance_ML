# TRIN (Trade Index) strategy


- 선물 매매 전략

① - TRIN 이 UBB 를 교차하면 BUY (new Position open)  
② - 이미 BUY position 이 open 되어 있으므로 open 된 position 유지  
③ - TRIN 이 UBB 에서 SMA 로 돌아오면 SELL. (Position 청산)  
④ - TRIN 이 LBB 를 교차하면 SELL (new Position open)  
⑤ - 이미 SELL position 이 open 되어 있으므로 open 된 position 유지  
⑥ - TRIN 이 UBB 에서 SMA 로 돌아오면 BUY. (Position 청산)  
⑦ - ① 과 동일  
⑧ - BUY position 이 open 된 후, TRIN 이 USB(Upper Stop-loss Band) 돌파하면, SELL 하고 loss 기장  
⑨ - ④ 와 동일
⑩ - SELL position 이 open 된 후, TRIN 이 LSB(Lower Stop-loss Band) 돌파하면, BUY 하고 loss 기장

**폭락장세등이 있을 경우 SMA 에 따라 움직이는 Bolinger Band 는 시세를 즉시 반영하지 못하므로 absolute stop loss 값을 정한다.**

### Bolinger Band 작성을 위한 sigma 계산 helper function

```python
def variance_calculator(series, series_average, win_len):   
    sma = win_len
    temp1 = series.subtract(series_average)           # 편차 계산 (a-b)
    temp2 = temp1.apply(lambda x: x**2)            # (a-b)^2
    temp3 = temp2.rolling(sma - 1).mean()            # sum((a-b)^2) / (sma-1)
    sigma = temp3.apply(lambda x: math.sqrt(x))  # standard deviation
    return sigma
```

상승 및 하락 주식의 수와 볼륨을 사용하여 TRIN 값을 찾습니다. 

TRIN(Trading Index) = $\frac{AD\ ratio}{AD\ volume\ ratio}$

AD ratio = $\frac{상승종목의\ 수}{하락종목의\ 수}$

AD volume ratio = $\frac{상승종목의\ 거래량}{하락종목의\ 거래량}$

```python
AD_ratio = Data['advancing'] / Data['declining']
AD_vol_ratio = adv_vol['Numbers of Stocks'].divide(dec_vol['Numbers of Stocks'])
```

```python
TRIN = AD_ratio / AD_vol_ratio
```

```python
Data['TRIN'] = TRIN

Data.head()
```

### TRIN 은 계산 공식의 특성에 의해 매도가 큰 날은 1 의 정수배로 커지고 (ex. 3) 매수가 큰 날은 소숫점 (ex. 0.3) 로 표시되는 속성이 있으므로 log 를 취해 normalize 해 준다

```python
Data['TRIN'] = Data['TRIN'].apply(lambda x: math.log(x))
```

## Backtesting

```python
sma = 22     # moving average window length
k = 1.2      # Bollinger band 작성시 sigma 의 몇배를 적용할 것인가 하는 상수
l = 1.5      # Bollinger band 에서 sigma 의 몇배에서 stop loss 할 것인지 정하는 상수
abs_SL = 25

def TRIN_PCR_Strategy(Data, sma=sma, k=k, l=l, abs_SL=abs_SL, TRIN_PCR="TRIN"):
    pro = 0               # profit 초기값
    flag = 1   # first transaction 을 하기 위한 flag. 거래는 LBB/UBB crossing over 에서만 시작
    buy_flag = False  # open 된 pisition 상태
    sell_flag = False
    transaction_start_price = 0   
    mtm = list()
    order_details = list()
    order = list()     # orders list : BUY/SELL/DO_nothing
    profit = list()
    buy_sell = list()
    stoploss = list()
    trade_cause = list()
    
    Data['mAvg'] = Data[TRIN_PCR].rolling(sma).mean()   # moving average of TRIN
    Data['previous'] = Data[TRIN_PCR].shift(1)
    sigma = variance_calculator(Data[TRIN_PCR], Data['mAvg'], sma)  # standard deviation
    
    k_sigma = k * sigma
    l_sigma = l * sigma
    
    Data['UBB'] = Data['mAvg'].add(k_sigma)        # Upper Bolinger Band
    Data['LBB'] = Data['mAvg'].subtract(k_sigma)  # Lower Bolinger Band
    Data['USL'] = Data['UBB'].add(l_sigma)           # Upper Stop Loss Band
    Data['LSL'] = Data['LBB'].subtract(l_sigma)     # Lower Stop Loss Band
    Data['order'] = pd.Series()    # list of orders - BUY/SELL/DO_nothing
    
    s = Data[TRIN_PCR].size           # total number of data points
    
    for i in range(s):

        pro = 0              # profit at each trade
        # 비교할 변수들
        future_cost = Data['future'][i]  #s&p500 선물 매입 비용
        TRIN      =   Data[TRIN_PCR][i]   #current TRIN ratio value
        previous  =   Data['previous'][i]  #previous day's TRIN ratio(for crossover check)
        LBB      =   Data['LBB'][i]        #lower bollinger band
        UBB      =   Data['UBB'][i]        #upper bollinger band
        mAvg     =   Data['mAvg'][i]       #moving average
        USL      =   Data['USL'][i]        #upper stoploss band
        LSL      =   Data['LSL'][i]        #lower stoploss band

        #주문하기 위해 저장한 부울 변수
        # TRIN 이 UBB 를 상향 돌파했는지 check (1, 2, 7)
        UBB_cross        =   (TRIN > UBB) and (previous < UBB)  
        # TRIN 이 LBB 를 하향 돌파했는지 check  (4, 5, 9)
        LBB_cross        =   (TRIN < LBB) and (previous > LBB)   
        # TRIN 이 SMA 를 상향돌파 했는지 check (6)
        mAvg_cross_up      =   (TRIN > mAvg) and (previous < mAvg)  
        # TRIN 이 SMA 를 하향돌파 했는지 check (3)
        mAvg_cross_down  =   (TRIN < mAvg) and (previous > mAvg)  
        # TRIN 이 USB 를 상향돌파 했는지 check (8)
        USL_cross        =   (TRIN > USL)  and  (previous < USL)  
        # TRIN 이 LSB 를 하향돌파 했는지 check (10)
        LSL_cross        =   (TRIN < LSL)  and  (previous > LSL)   

        # TRIN 이 UBB 를 상향돌파하고 open 된 BUY position 이 없으면 매수 주문 (1, 2, 7)
        if(UBB_cross and (not buy_flag) and flag ==1): 
            flag = 0             # open 된 positon 존재
            buy_flag = True      # BUY position    
            sell_flag = False
            transaction_start_price = future_cost   # 주문시의 s&p500 선물 매수 가격
            order_details = [1,"Buy" , "UBB crossed" , "0" , "position taken"]

        # TRIN 이 LBB 를 하향돌파하고 open 된 SELL position 이 없으면 매도 주문 (4, 5, 9)
        elif (LBB_cross and (not sell_flag) and flag ==1): 
            flag = 0  
            sell_flag = True       # SELL position
            buy_flag = False
            transaction_start_price = future_cost    # 주문시의 s&p500 선물 매도 가격
            order_details = [-1,"Sell" , "LBB crossed" , "0" , "position taken"]

        # TRIN 이 SMA 를 상향돌파하면 매수 주문 내고 position 청산 (6)
        elif (mAvg_cross_up and flag==0 and (not buy_flag)) : 
            flag = 1          # open 된 position 없음
            buy_flag = False 
            sell_flag = False
            # 청산이익 = 당일선물가 - 최초 선물매도가
            pro = future_cost - transaction_start_price   
            order_details = [1,"Buy" , "mAvg crossed" , "0" , "position closed"]

        # TRIN 이 LSB 를 하향 돌파하면 매수 주문내고 position 청산 (10)
        elif(LSL_cross and flag == 0 and (not buy_flag)): 
            flag = 1
            buy_flag = False
            sell_flag = False
            # 청산이익 = 당일선물가 - 최초 선물매도가
            pro = future_cost - transaction_start_price       
            order_details = [1,"Buy" , "LSB crossed" , 
                             "stoploss executed" , "position closed"]

        elif((future_cost - transaction_start_price) > abs_SL and \
             flag == 0 and (not buy_flag)): 
            # open 된 position 있고 BUY position 이 아니면서 TRIN 이 abs stoploss 값을 
            # 하향돌파하면 지수가 폭등한 것이므로 매수 주문을 하여 SELL position 이 있을 경우 청산
            flag = 1
            buy_flag = False
            sell_flag = False
            pro = future_cost - transaction_start_price
            order_details = [1,"Buy" , "LSB crossed" , 
                             "stoploss executed abs" , "position closed"]

        # TRIN 이 SMA 를 하향돌파하면 매도 주문 내고 position 청산 (3)
        elif (mAvg_cross_down and flag==0 and (not sell_flag)):  
            flag = 1
            sell_flag = False
            buy_flag = False
            # 청산이익 = 최초 선물매수가 - 당일선물가격
            pro = -(future_cost - transaction_start_price)              
            order_details = [-1,"Sell" , "mAvg crossed (h to l)" , 
                             "0" , "position closed"]

        # TRIN 이 USB 를 돌파하면 매도 주문내고 position 청산 (8)
        elif(USL_cross and flag==0 and (not sell_flag)):  
            flag = 1
            sell_flag = False
            buy_flag = False
            pro = -(future_cost - transaction_start_price)
            order_details = [-1,"Sell" , "USB crossed" , 
                             "stoploss executed" , "position closed"]

        elif((-future_cost + transaction_start_price) > abs_SL and \
                     flag==0 and (not sell_flag)): 
            # open 된 position 있고 SELL position 이 아니면서 TRIN 이 abs stoploss 값을 
            # 상향돌파하면 지수가 폭락한 것이므로 매도 주문을 하여 BUY position 이 있을 경우 청산
            flag = 1
            sell_flag = False
            buy_flag = False
            pro = -(future_cost - transaction_start_price)
            order_details = [-1,"Sell" , "USB crossed" , 
                             "stoploss executed_abs" , "position closed"]

        else:
            if(buy_flag==0 and sell_flag==0): tempo = "0"  # no active trades
            else:
                if(buy_flag==1 and sell_flag==0): 
                    tempo = (Data['future'][i] - \
                            transaction_start_price) * 500  # 500: S&P500 선물 계약단위 
                if(buy_flag==0 and sell_flag==1): 
                    tempo= (-Data['future'][i] + \
                            transaction_start_price) * 500
            order_details = [0,"No trade" , "no trade" , "0" , tempo]

        profit.append(pro)                 # profit / loss 기록
        order.append(order_details[0])     # 1: BUY, -1: SELL, 0: No trade
        buy_sell.append(order_details[1])    # "Buy", "Sell", "No trade"
        trade_cause.append(order_details[2])  # "USB crossed", etc
        stoploss.append(order_details[3])     # "stoploss execute", "0"
        mtm.append(order_details[4])        # "position taken", "position closed", tempo
        
    Data['placed_order'] = pd.Series(order)
    Data['cost'] = -(Data['placed_order'].multiply(Data['future'])) * 500
    Data['out']  = Data['cost'].cumsum()        # cash out
    Data['buy_sell'] = pd.Series(buy_sell)
    Data['profit'] = -pd.Series(profit) * 500
    Data['stoploss'] = pd.Series(stoploss)
    Data['trade_cause'] = pd.Series(trade_cause)
    Data['mtm'] = pd.Series(mtm)
    
    return Data
```

```python
Data = TRIN_PCR_Strategy(Data)
```
