# 다변수 선형회귀 모델

- 여러 stock market 의 historical data 를 이용하여 SPY 를 trading 하는 model 개발  


- SPY - 1993년에 만들어진 S&P500 추종 ETF(Exchange Traded Fund, 상장지수펀드)

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as data_reader
from pandas.plotting import scatter_matrix
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings(action='ignore')
```

### 모든 주식 시장의 데이터를 DataFrame으로 가져오기

```python
spy = data_reader.DataReader("SPY", data_source="yahoo")
spy.head()
```

```python
nasdaq = data_reader.DataReader("^IXIC", data_source="yahoo")
sp500 = data_reader.DataReader("^GSPC", data_source="yahoo")
dji = data_reader.DataReader("^DJI", data_source="yahoo")

# DAX(독일), CAC(프랑스)
daxi = data_reader.DataReader("^GDAXI", data_source="yahoo")
cac40 = data_reader.DataReader("^FCHI", data_source="yahoo")

# aord(호주), N225(니케이)
aord = data_reader.DataReader("^AORD", data_source="yahoo")
nikkei = data_reader.DataReader("^N225", data_source="yahoo")
```

## Step1: Data Munging

- timezone 을 감안하여, 적절한 stock market data 를 추출하여 계산. timezone 이 다른 시간대의 시장 정보를 이용하여 다른 시간대의 시장 예측

    - US Market  
        - SPY     : Open – 어제 시초가  
        - SP500  : Open – 어제 시초가 
        - Nasdaq: Open – 어제 시초가  
        - DJI      : Open – 어제 시초가  
    - EU Market  
        - CAC40:  Open – 어제 시초가 
        - DAXI  :  Open – 어제 시초가  
    - ASIAN Market  
        - AORD:  Close – 당일 시초가 
        - HIS   :   Close – 당일 시초가    
        - Nikkei:  Close - 당일 시초가   
    
    
- 선형회귀 모델로 알아내고자 하는 target 은 **내일의 시초가**의 상승, 하락 이다.

- spy 를 포함한 각 index 의 (당일 Open - 전일 Open) 으로 내일의 Open 이 상승일지, 하락일지 하루 전 예측하는 모델 작성

```python
# empty DataFrame 생성
df = pd.DataFrame(index=spy.index)

# SPY (내일 시초가 - 당일 시초가) : 내일 시초가의 상승, 하락 예측
df['spy_target'] = spy['Open'].shift(-1) - spy['Open']  
df.head()
```

```python
# US, EU market의 당일 시초가 등락(당일시초가 - 전일시초가)
df['spy']    = spy['Open'] - spy['Open'].shift(1)  
df['sp500']  = sp500["Open"] - sp500['Open'].shift(1)
df['nasdaq'] = nasdaq['Open'] - nasdaq['Open'].shift(1)
df['dji']    = dji['Open'] - dji['Open'].shift(1)
df['cac40']  = cac40['Open'] - cac40['Open'].shift(1)
df['daxi']   = daxi['Open'] - daxi['Open'].shift(1)
df.head()
```

```python
# ASIAN market의 당일 등락(당일종가 - 당일시초가)
df['aord']   = aord['Close'] - aord['Open']
df['nikkei'] = nikkei['Close'] - nikkei['Open']
df.head()
```

```python
# SPY 실제 당일 시초가
df['spy_price'] = spy['Open']
df.head()
```

- 각 국의 휴일등이 달라 NaN 이 있을 수 있으므로 NaN check
- fillna(method='ffill') 으로 채우고 나머지는 drop

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill', inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
df.to_csv("data/indicepanel.csv")
```

## Step 4 : Model Build and Prediction

- SPY의 당일 등락 예측

- sklearn 의 `linear_model.LinearRegression` 사용

```python
train.head()
```

```python
X_train = train.iloc[:, 1:-1]
y_train = train.iloc[:, 0]

X_test = test.iloc[:, 1:-1]
y_test = test.iloc[:, 0]

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
lr = linear_model.LinearRegression()

lr.fit(X_train, y_train)
```

```python
# The coefficients
print('Coefficients: \n', lr.coef_)
print('Intercept: \n', lr.intercept_)
```

```python
y_pred = lr.predict(X_test)

# R2 계산 
print("R2 : {:.2f}".format(r2_score(y_test, y_pred)))
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('spy target')
plt.ylabel('Prediction');
```

### 다른 주식 시장의 등락을 이용하여 다변수 선형회귀 모델로 SPY를 예측하기 어려움을 알 수 있다.

## 전략 비교
### Signal Based 전략
- 다음날 시초가 상승할 것 (>0) 으로 예상되면 1 주 buy, 하락할 것 (<0) 으로 예상되면 1 주 sell  

### 분할 매수 only 전략
- 매일 한주씩 매수하는 Buy-Only 전략과 비교

```python
test = df.iloc[-650:, :]
print(test.shape)
```

```python
test['predicted']  = lr.predict(X_test)
test.head()
```

```python
# signal based 전략
test['order_signal'] = [1 if sig > 0 else -1 for sig in test['predicted']]
# 보유 주식 잔고
test['stock_balance_signal'] = test['order_signal'].cumsum()
# spy 등락에 따른 보유 주식 평가 손익
test['profit-signal'] = test['spy'] * test['stock_balance_signal'] 
# 누적 손익
test['wealth_signal'] = test['profit-signal'].cumsum()
```

```python
# buy-only 전략
test['order_buy_only'] = 1
# 보유 주식 잔고
test['stock_balance_buy_only'] = test['order_buy_only'].cumsum()
# spy 등락에 따른 보유 주식 평가 손익
test['profit-buy_only'] = test['spy'] * test['stock_balance_buy_only'] 
# 누적 손익
test['wealth_buy_only'] = test['profit-buy_only'].cumsum()
```

```python
test.tail()
```

```python
s = test.index[0].strftime('%Y-%m-%d')
e = test.index[-1].strftime('%Y-%m-%d')

print(f"{s} ~ {e} 기간에서 : ")
print(f"Signal Based 전략을 따른 경우 test set 에서 예측되는 total profit: ${test['profit-signal'].sum():.2f}")
print(f"Greedy 전략을 따른 경우 test set 에서 예측되는 total profit: ${test['profit-buy_only'].sum():.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.title('Performance of Strategy in Train')
plt.plot(test['wealth_signal'].values, color='green', label='Signal based strategy')
plt.plot(test['wealth_buy_only'].values, color='red', label='Buy Only strategy')

plt.legend()
```

### 해당 기간 동안은 spy 가 지속적 상승을 했으므로 보유 주식 수를 늘리는 전략이 유효했으나, spy 하락 기간 동안에는 Buy Only 전략의 손실이 Signal Based 전략 보다 훨씬 컸음을 주목
