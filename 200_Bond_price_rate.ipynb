# 채권 가격과 이자율 관계 시각화

액면가: 1,000달러  
쿠폰 금리: 5% (연간 50달러 지급)  
만기: 10년  

- 시장 이자율이 **5%**일 때 투자자는 채권이 액면가(1,000달러)로 거래될 것이라 기대  
- 시장 이자율이 **6%**로 상승하면 채권의 고정된 이자 지급액(50달러)은 매력적이지 않게 되므로 채권 가격은 할인  
- 시장 이자율이 **4%**로 하락하면 채권의 이자 지급액(50달러)은 매력적으로 보이게 되므로 채권 가격의 프리미엄 상승  

```python
import numpy as np
import matplotlib.pyplot as plt

# 채권의 기본 정보
face_value = 1000  # 액면가 (1,000 달러)
coupon_rate = 0.05  # 쿠폰 금리 (5%)
years = 10  # 만기 (10년)

# 다양한 시장 이자율 (0%에서 10%까지)
interest_rates = np.linspace(0.01, 0.10, 100)
```

```python
# 채권 가격 계산 함수 - 주어진 시장 이자율에서 채권 가격을 계산
def bond_price(face_value, coupon_rate, interest_rate, years):
    coupon_payment = face_value * coupon_rate  # 연간 쿠폰 지급액
    price = sum([coupon_payment / (1 + interest_rate) ** t for t in range(1, years + 1)]) \
            + face_value / (1 + interest_rate) ** years
    return price

# 각 이자율에 대한 채권 가격 계산
bond_prices = [bond_price(face_value, coupon_rate, rate, years) for rate in interest_rates]

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(interest_rates * 100, bond_prices, label="Bond Price", color="blue")
plt.xlabel("Market Interest Rate (%)")
plt.ylabel("Bond Price (USD)")
plt.title("Relationship Between Bond Price and Interest Rate")
plt.grid(True)
plt.axhline(face_value, color="red", linestyle="--", label="Face Value ($1000)")
plt.legend()
plt.show()
```
